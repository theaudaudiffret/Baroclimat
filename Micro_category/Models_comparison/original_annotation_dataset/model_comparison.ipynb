## 4 bits Quantization comparison


```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_lora_4_bits = pd.read_csv('./predictions_lora_new/LORA_METRICS_NEW.csv')
df_base_model_4_bits = pd.read_csv('./predictions_base_model/METRICS_BASE_NEW.csv')



# Define the metrics to plot
metrics = ['Precision', 'Recall', 'F1_score', 'Top 1 Accuracy', 'Top 3 Accuracy']
x = np.arange(len(metrics))  # the label locations
width = 0.35  # the width of the bars

# Create a plot for each model
for model in df_lora_4_bits.index:
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot the metrics for the lora model
    ax.bar(x - width/1.85, df_lora_4_bits.loc[model, metrics], width, label='Lora Model')
    
    # Plot the metrics for the base model
    ax.bar(x + width/1.85, df_base_model_4_bits.loc[model, metrics], width, label='Base Model')
    
    # Set plot title and labels
    ax.set_title(f'Metrics Comparison for {df_lora_4_bits.loc[model, "source"]}')

    ax.set_xlabel('Metrics')
    ax.set_ylabel('Values')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()
    
    # Show the plot
    plt.show()
```

# Metrics for the selected best model

```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_lora_best = pd.read_csv('../Final_model/resultats_lora.csv')
df_base_best = pd.read_csv('../Final_model/resultats_base.csv')
df_benchmark = pd.read_csv('../Final_model/Benchmark.csv')



# Define the metrics to plot
metrics = ['Precision', 'Recall', 'F1_score', 'Top 1 Accuracy', 'Top 3 Accuracy']
x = np.arange(len(metrics))  # the label locations
width = 0.20  # the width of the bars

# Create a plot for each model
for model in df_lora_best.index:
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot the metrics for the lora model
    ax.bar(x - width*1.05, df_lora_best.loc[model, metrics], width, label='Lora-Model')
    
    # Plot the metrics for the base model
    ax.bar(x, df_base_best.loc[model, metrics], width, label='Base-model')
    
    ax.bar(x + width*1.05, df_benchmark.loc[model, metrics], width, label='Benchmark semi-supervisé')

    
    # Set plot title and labels
    ax.set_title(f'Metrics Comparison for {df_lora_best.loc[model, "source"]}')

    ax.set_xlabel('Metrics')
    ax.set_ylabel('Values')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()
    

    # Show the plot
    plt.show()
```
