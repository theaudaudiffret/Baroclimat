# Inference on the new dataset


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Define the metrics to be compared
df = pd.read_csv('./results.csv')
metrics = ['Precision', 'Recall', 'F1_score', 'Top 1 Accuracy']
# Create a single figure and axis for all models
x = np.arange(len(metrics))  # the label locations
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.15  # Width of each bar
n_models = len(df)
for i, model in enumerate(df.index):
    offset = -width * (n_models - 1) / 2 + i * width
    ax.bar(x + offset, df.loc[model, metrics], width, label=df.loc[model, 'source'])

ax.set_title('Metrics Comparison for models on the new dataset')
ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

plt.show()
```
